## Задача

    f(&x) \rightarrow min \\

## Градиентый спуск
Небольшие шаги в направлении антиградиента из текущей точки

```python
def f(x):
    return x ** 2

def grad_f(x):
    return 2 * x

def alpha(n):
    return 5. / (1 + n/3)
```

```python
x = np.linspace(-3, 3, 1000)
y = f(x)
```

```python
%matplotlib inline

plt.figure(figsize=(12, 8))
plt.style.use('ggplot')
plt.title("$y = x^2$")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y);
```

```python
def grad_descent(f, grad_f, alpha, x_0, delta):
    n = 0
    x = x_0 + 2 * delta
    path = []
    while abs(x_0 - x) > delta and n < 100:
        x_0 = x
        grad = grad_f(x)
        x = x - alpha(n) * grad / abs(grad)
        n += 1
        path.append(x)
        
    return path
```

```python
x_0 = 3
path = grad_descent(f, grad_f, alpha, x_0, 1e-4)
```

```python
plt.figure(figsize=(12, 8))
plt.style.use('ggplot')
plt.title("$y = x^2$")
plt.plot(x, y);
for x_next in path:
    plt.plot([x_0, x_next], [f(x_0), f(x_next)])
    x_0 = x_next
plt.xlabel("x")
plt.ylabel("y");
```
